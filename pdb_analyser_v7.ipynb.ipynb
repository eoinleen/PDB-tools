````python
"""
=============================================================================
                     COMPREHENSIVE PDB STRUCTURE ANALYZER
                        WITH MISSING ATOM DETECTION
=============================================================================

DESCRIPTION:
This script provides comprehensive analysis and validation of Protein Data Bank
(PDB) files with enhanced parsing, structural validation, sequence extraction,
missing atom detection, and data visualization capabilities.

MAIN FEATURES:
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

📊 STRUCTURAL ANALYSIS:
   • Parses ATOM and HETATM records with comprehensive field extraction
   • Validates atom numbering continuity and identifies gaps
   • Analyzes residue numbering per chain and detects missing residues
   • Extracts coordinate ranges (X, Y, Z) and B-factor distributions
   • Identifies alternative conformations and insertion codes
   • Locates structural markers (TER, END records)

🔍 MISSING ATOM DETECTION:
   • Compares actual atoms vs expected atoms for each residue type
   • Identifies truncated side chains (e.g., lysine → alanine)
   • Detects missing backbone atoms (N, CA, C, O)
   • Reports completeness percentage for each residue
   • Highlights common mutations and truncations

🔍 DATA VALIDATION:
   • Comprehensive error checking for malformed PDB lines
   • Chain-by-chain statistical analysis
   • Residue composition analysis per chain
   • Coordinate boundary validation
   • Occupancy and temperature factor analysis

📈 VISUALIZATION:
   • 3D structure projection (X-Y coordinates with chain coloring)
   • B-factor distribution histograms
   • Residue count comparison across chains
   • Top 10 most frequent atom types
   • Chain length and composition charts
   • Missing atom statistics visualization

🧬 SEQUENCE EXTRACTION:
   • Automatic protein sequence extraction per chain
   • Three-letter to single-letter amino acid conversion
   • Handles modified residues (MSE, SEC, PYL)
   • FASTA format export with detailed headers
   • Non-standard residue detection and reporting
   • Sequence length and range statistics

🛠️ FILE PROCESSING:
   • Atom renumbering functionality for non-sequential structures
   • Enhanced PDB line formatting with all standard fields
   • Automatic file upload in Google Colab environment
   • Output generation for corrected PDB files

⚡ IMMEDIATE EXECUTION:
   • No function calls required - runs automatically upon execution
   • Real-time analysis and reporting
   • Interactive file upload interface
   • Comprehensive console output with progress indicators

USAGE:
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
Simply run this script in a Jupyter/Colab environment. It will:
1. Prompt for PDB file upload
2. Perform comprehensive structural analysis
3. Check for missing atoms in amino acid side chains
4. Generate validation reports and statistics
5. Create visualization plots
6. Extract and export protein sequences in FASTA format
7. Offer atom renumbering if structural issues are detected

OUTPUT FILES:
   • [filename]_sequences.fasta - Protein sequences for all chains
   • [filename]_missing_atoms.txt - Detailed missing atom report
   • renumbered.pdb - Corrected PDB file (if renumbering is performed)

REQUIREMENTS:
   • pandas, matplotlib, numpy (data processing and visualization)
   • Google Colab environment (for file upload functionality)
   • Standard Python libraries: collections, warnings

AUTHOR: Enhanced PDB Analysis Script
VERSION: 7.0 - Missing Atom Detection Edition
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
"""

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter, defaultdict
import warnings

# Upload the PDB file
from google.colab import files
print("📤 Please upload your PDB file:")
uploaded = files.upload()
filename = list(uploaded.keys())[0]

print(f"📄 Analyzing PDB file: {filename}")

# Read all lines from the file
with open(filename, 'r') as f:
    lines = f.readlines()

print(f"📏 Total lines in file: {len(lines)}")
print("=" * 60)

# Define expected atoms for each amino acid residue
EXPECTED_ATOMS = {
    'ALA': ['N', 'CA', 'C', 'O', 'CB'],
    'ARG': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'CD', 'NE', 'CZ', 'NH1', 'NH2'],
    'ASN': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'OD1', 'ND2'],
    'ASP': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'OD1', 'OD2'],
    'CYS': ['N', 'CA', 'C', 'O', 'CB', 'SG'],
    'GLU': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'CD', 'OE1', 'OE2'],
    'GLN': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'CD', 'OE1', 'NE2'],
    'GLY': ['N', 'CA', 'C', 'O'],
    'HIS': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'ND1', 'CD2', 'CE1', 'NE2'],
    'ILE': ['N', 'CA', 'C', 'O', 'CB', 'CG1', 'CG2', 'CD1'],
    'LEU': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'CD1', 'CD2'],
    'LYS': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'CD', 'CE', 'NZ'],
    'MET': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'SD', 'CE'],
    'PHE': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'CD1', 'CD2', 'CE1', 'CE2', 'CZ'],
    'PRO': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'CD'],
    'SER': ['N', 'CA', 'C', 'O', 'CB', 'OG'],
    'THR': ['N', 'CA', 'C', 'O', 'CB', 'OG1', 'CG2'],
    'TRP': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'CD1', 'CD2', 'NE1', 'CE2', 'CE3', 'CZ2', 'CZ3', 'CH2'],
    'TYR': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'CD1', 'CD2', 'CE1', 'CE2', 'CZ', 'OH'],
    'VAL': ['N', 'CA', 'C', 'O', 'CB', 'CG1', 'CG2'],
    # Common modified residues
    'MSE': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'SE', 'CE'],  # Selenomethionine
    'SEC': ['N', 'CA', 'C', 'O', 'CB', 'SEG'],  # Selenocysteine
    'PYL': ['N', 'CA', 'C', 'O', 'CB', 'CG', 'CD', 'CE', 'NZ', 'CH', 'CZ']  # Pyrrolysine
}

# Backbone atoms (essential for all residues)
BACKBONE_ATOMS = ['N', 'CA', 'C', 'O']

# Enhanced parsing function with more PDB fields
def parse_pdb_atoms(lines):
    atoms = []
    for i, line in enumerate(lines):
        if line.startswith(('ATOM', 'HETATM')) and len(line) >= 54:
            try:
                atom = {
                    'record_type': line[0:6].strip(),
                    'atom_number': int(line[6:11]),
                    'atom_name': line[12:16].strip(),
                    'alt_loc': line[16].strip(),
                    'residue_name': line[17:20].strip(),
                    'chain_id': line[21].strip(),
                    'residue_number': int(line[22:26]),
                    'insertion_code': line[26].strip(),
                    'x': float(line[30:38]),
                    'y': float(line[38:46]),
                    'z': float(line[46:54]),
                    'occupancy': float(line[54:60]) if len(line) > 54 and line[54:60].strip() else 1.0,
                    'temp_factor': float(line[60:66]) if len(line) > 60 and line[60:66].strip() else 0.0,
                    'element': line[76:78].strip() if len(line) > 76 else '',
                    'line_number': i + 1
                }
                atoms.append(atom)
            except (ValueError, IndexError) as e:
                continue
    return pd.DataFrame(atoms)

def check_missing_atoms(df):
    """Check for missing atoms in amino acid residues."""
    missing_atoms_report = []
    chain_summaries = {}

    print("\n" + "="*60)
    print("MISSING ATOM ANALYSIS")
    print("="*60)

    for chain_id, chain_group in df.groupby('chain_id'):
        chain_name = chain_id if chain_id else 'blank'
        print(f"\n🔍 Analyzing Chain {chain_name}:")

        chain_stats = {
            'total_residues': 0,
            'complete_residues': 0,
            'missing_backbone': 0,
            'missing_sidechain': 0,
            'truncated_residues': 0
        }

        # Group by residue
        for (res_num, res_name), residue_group in chain_group.groupby(['residue_number', 'residue_name']):
            if res_name not in EXPECTED_ATOMS:
                continue  # Skip non-standard residues

            chain_stats['total_residues'] += 1
            expected_atoms = set(EXPECTED_ATOMS[res_name])
            actual_atoms = set(residue_group['atom_name'].tolist())

            # Remove alternative location indicators for comparison
            actual_atoms_clean = {atom.split('.')[0] for atom in actual_atoms}
            missing_atoms = expected_atoms - actual_atoms_clean

            # Check backbone completeness
            missing_backbone = set(BACKBONE_ATOMS) - actual_atoms_clean
            missing_sidechain = missing_atoms - set(BACKBONE_ATOMS)

            if not missing_atoms:
                chain_stats['complete_residues'] += 1
            else:
                # Record missing atoms
                missing_info = {
                    'chain': chain_name,
                    'residue_number': res_num,
                    'residue_name': res_name,
                    'missing_atoms': list(missing_atoms),
                    'missing_backbone': list(missing_backbone),
                    'missing_sidechain': list(missing_sidechain),
                    'completeness': (len(actual_atoms_clean) / len(expected_atoms)) * 100,
                    'actual_atoms': list(actual_atoms_clean),
                    'expected_atoms': list(expected_atoms)
                }
                missing_atoms_report.append(missing_info)

                if missing_backbone:
                    chain_stats['missing_backbone'] += 1
                if missing_sidechain:
                    chain_stats['missing_sidechain'] += 1

                # Check for common truncations
                if res_name in ['LYS', 'ARG', 'GLU', 'GLN', 'MET', 'LEU', 'ILE', 'VAL', 'PHE', 'TYR', 'TRP', 'HIS']:
                    if len(missing_sidechain) > 0:
                        chain_stats['truncated_residues'] += 1

        chain_summaries[chain_name] = chain_stats

        # Print chain summary
        if chain_stats['total_residues'] > 0:
            completeness = (chain_stats['complete_residues'] / chain_stats['total_residues']) * 100
            print(f"   📊 Residues: {chain_stats['total_residues']}")
            print(f"   ✅ Complete: {chain_stats['complete_residues']} ({completeness:.1f}%)")
            print(f"   ❌ Missing backbone atoms: {chain_stats['missing_backbone']}")
            print(f"   ⚠️  Missing sidechain atoms: {chain_stats['missing_sidechain']}")
            print(f"   ✂️  Truncated residues: {chain_stats['truncated_residues']}")

    return missing_atoms_report, chain_summaries

def analyze_missing_patterns(missing_atoms_report):
    """Analyze patterns in missing atoms."""
    print("\n" + "="*60)
    print("MISSING ATOM PATTERNS")
    print("="*60)

    if not missing_atoms_report:
        print("✅ No missing atoms detected!")
        return

    # Common truncation patterns
    truncation_patterns = defaultdict(list)
    residue_completeness = defaultdict(list)

    for record in missing_atoms_report:
        res_name = record['residue_name']
        missing = record['missing_atoms']
        completeness = record['completeness']

        residue_completeness[res_name].append(completeness)

        # Identify truncation patterns
        if res_name == 'LYS' and 'NZ' in missing:
            if set(missing) == {'CD', 'CE', 'NZ'}:
                truncation_patterns['LYS→LEU'].append(record)
            elif set(missing) == {'CG', 'CD', 'CE', 'NZ'}:
                truncation_patterns['LYS→VAL'].append(record)
            elif set(missing) == {'CB', 'CG', 'CD', 'CE', 'NZ'}:
                truncation_patterns['LYS→GLY'].append(record)

        elif res_name == 'ARG' and 'NE' in missing:
            if 'CZ' in missing and 'NH1' in missing and 'NH2' in missing:
                truncation_patterns['ARG→truncated'].append(record)

        elif res_name == 'GLU' and 'OE1' in missing and 'OE2' in missing:
            truncation_patterns['GLU→truncated'].append(record)

        elif res_name == 'GLN' and 'OE1' in missing and 'NE2' in missing:
            truncation_patterns['GLN→truncated'].append(record)

        elif res_name == 'MET' and 'SD' in missing and 'CE' in missing:
            truncation_patterns['MET→truncated'].append(record)

        elif res_name in ['PHE', 'TYR', 'TRP'] and any(atom.startswith('C') for atom in missing):
            truncation_patterns[f'{res_name}→truncated'].append(record)

    # Print truncation patterns
    if truncation_patterns:
        print("\n🔍 Detected truncation patterns:")
        for pattern, records in truncation_patterns.items():
            print(f"   {pattern}: {len(records)} occurrences")
            for record in records[:3]:  # Show first 3 examples
                print(f"     Chain {record['chain']}, Residue {record['residue_number']}")
            if len(records) > 3:
                print(f"     ... and {len(records) - 3} more")

    # Print residue completeness statistics
    print("\n📊 Residue completeness statistics:")
    for res_name, completeness_list in residue_completeness.items():
        avg_completeness = np.mean(completeness_list)
        min_completeness = np.min(completeness_list)
        incomplete_count = len([c for c in completeness_list if c < 100])
        total_count = len(completeness_list)

        print(f"   {res_name}: {avg_completeness:.1f}% avg, {min_completeness:.1f}% min, "
              f"{incomplete_count}/{total_count} incomplete")

def write_missing_atoms_report(missing_atoms_report, filename_base):
    """Write detailed missing atoms report to file."""
    report_filename = f"{filename_base}_missing_atoms.txt"

    with open(report_filename, 'w') as f:
        f.write("MISSING ATOMS REPORT\n")
        f.write("="*50 + "\n\n")

        if not missing_atoms_report:
            f.write("No missing atoms detected!\n")
            return report_filename

        f.write(f"Total residues with missing atoms: {len(missing_atoms_report)}\n\n")

        # Group by chain
        by_chain = defaultdict(list)
        for record in missing_atoms_report:
            by_chain[record['chain']].append(record)

        for chain_name, records in by_chain.items():
            f.write(f"CHAIN {chain_name}\n")
            f.write("-" * 20 + "\n")

            for record in records:
                f.write(f"Residue {record['residue_name']}{record['residue_number']}:\n")
                f.write(f"  Completeness: {record['completeness']:.1f}%\n")
                f.write(f"  Missing atoms: {', '.join(record['missing_atoms'])}\n")
                if record['missing_backbone']:
                    f.write(f"  Missing backbone: {', '.join(record['missing_backbone'])}\n")
                if record['missing_sidechain']:
                    f.write(f"  Missing sidechain: {', '.join(record['missing_sidechain'])}\n")
                f.write(f"  Present atoms: {', '.join(sorted(record['actual_atoms']))}\n")
                f.write("\n")
            f.write("\n")

    print(f"📄 Missing atoms report written to: {report_filename}")
    return report_filename

# Parse atom records
df = parse_pdb_atoms(lines)

if df.empty:
    print("❌ No valid ATOM/HETATM records found!")
else:
    # Basic statistics
    print(f"✅ Total ATOM/HETATM records: {len(df)}")
    print(f"✅ Record types: {dict(df['record_type'].value_counts())}")
    print(f"✅ Total chains: {df['chain_id'].nunique()}")
    print(f"✅ Chain IDs: {sorted(df['chain_id'].unique())}")
    print(f"✅ Total residues: {df.groupby(['chain_id', 'residue_number']).ngroups}")

    # Locate TER and END records
    ter_lines = [i + 1 for i, line in enumerate(lines) if line.startswith('TER')]
    end_lines = [i + 1 for i, line in enumerate(lines) if line.startswith('END')]

    print(f"✅ TER records at lines: {ter_lines if ter_lines else 'None'}")
    print(f"✅ END records at lines: {end_lines if end_lines else 'None'}")

    # Atom number continuity check
    df['atom_diff'] = df['atom_number'].diff()
    non_seq_atoms = df[df['atom_diff'] != 1].copy()
    print(f"✅ Atom numbers sequential: {'Yes' if non_seq_atoms.empty else 'No'}")

    # Show atom number discontinuities if any
    if not non_seq_atoms.empty:
        print(f"\n❌ Found {len(non_seq_atoms)} breaks in atom numbering:")
        print(non_seq_atoms[['atom_number', 'atom_diff', 'line_number', 'chain_id',
                            'residue_number', 'atom_name']].head(10))

    # Check for alternative locations
    alt_locs = df[df['alt_loc'] != '']
    if not alt_locs.empty:
        print(f"\n⚠️  Found {len(alt_locs)} atoms with alternative locations")
        print(f"Alt location indicators: {sorted(alt_locs['alt_loc'].unique())}")

    # Coordinate ranges
    print(f"\n📊 Coordinate ranges:")
    print(f"  X: {df['x'].min():.3f} to {df['x'].max():.3f}")
    print(f"  Y: {df['y'].min():.3f} to {df['y'].max():.3f}")
    print(f"  Z: {df['z'].min():.3f} to {df['z'].max():.3f}")

    if df['temp_factor'].std() > 0:
        print(f"📊 B-factor range: {df['temp_factor'].min():.2f} to {df['temp_factor'].max():.2f}")

    # Residue number check per chain
    print("\n" + "="*60)
    print("CHAIN ANALYSIS")
    print("="*60)

    for chain_id, group in df.groupby('chain_id'):
        chain_name = chain_id if chain_id else '(blank)'
        print(f"\n--- Chain {chain_name} ---")

        residues = group['residue_number'].drop_duplicates().sort_values().reset_index(drop=True)

        # Find gaps in residue numbering
        gaps = []
        for i in range(1, len(residues)):
            if residues.iloc[i] - residues.iloc[i-1] > 1:
                gaps.append((residues.iloc[i-1], residues.iloc[i]))

        atom_count = len(group)
        residue_count = len(residues)

        print(f"Atoms: {atom_count}, Residues: {residue_count}")
        print(f"Residue range: {residues.min()} to {residues.max()}")

        if gaps:
            print(f"❌ Gaps in residue numbering: {gaps}")
        else:
            print("✅ Residue numbers sequential")

        # Most common residues
        common_residues = group['residue_name'].value_counts().head(3)
        print(f"Top residues: {dict(common_residues)}")

    # Check for missing atoms
    missing_atoms_report, chain_summaries = check_missing_atoms(df)

    # Analyze missing atom patterns
    analyze_missing_patterns(missing_atoms_report)

    # Write missing atoms report
    base_name = filename.replace('.pdb', '').replace('.PDB', '')
    write_missing_atoms_report(missing_atoms_report, base_name)

    # Create visualizations
    print("\n📈 Creating visualizations...")

    # Determine subplot layout based on data availability
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))

    # 3D scatter plot of atoms (projected to 2D)
    colors = df['chain_id'].astype('category').cat.codes
    axes[0,0].scatter(df['x'], df['y'], c=colors, alpha=0.6, s=1)
    axes[0,0].set_xlabel('X coordinate')
    axes[0,0].set_ylabel('Y coordinate')
    axes[0,0].set_title('Atom positions (X-Y projection)')

    # B-factor distribution
    if df['temp_factor'].std() > 0:
        axes[0,1].hist(df['temp_factor'], bins=50, alpha=0.7)
        axes[0,1].set_xlabel('B-factor')
        axes[0,1].set_ylabel('Count')
        axes[0,1].set_title('B-factor distribution')
    else:
        axes[0,1].text(0.5, 0.5, 'No B-factor variation', ha='center', va='center',
                      transform=axes[0,1].transAxes)
        axes[0,1].set_title('B-factor distribution')

    # Residue counts per chain
    chain_counts = df.groupby('chain_id')['residue_number'].nunique()
    axes[0,2].bar(range(len(chain_counts)), chain_counts.values)
    axes[0,2].set_xticks(range(len(chain_counts)))
    axes[0,2].set_xticklabels([f"Chain {c}" if c else "Chain (blank)"
                              for c in chain_counts.index])
    axes[0,2].set_ylabel('Number of residues')
    axes[0,2].set_title('Residues per chain')

    # Atom type distribution
    atom_types = df['atom_name'].value_counts().head(10)
    axes[1,0].barh(range(len(atom_types)), atom_types.values)
    axes[1,0].set_yticks(range(len(atom_types)))
    axes[1,0].set_yticklabels(atom_types.index)
    axes[1,0].set_xlabel('Count')
    axes[1,0].set_title('Top 10 atom types')

    # Missing atoms per chain
    chain_names = list(chain_summaries.keys())
    missing_counts = [chain_summaries[chain]['missing_backbone'] + chain_summaries[chain]['missing_sidechain']
                     for chain in chain_names]

    axes[1,1].bar(range(len(chain_names)), missing_counts)
    axes[1,1].set_xticks(range(len(chain_names)))
    axes[1,1].set_xticklabels([f"Chain {c}" for c in chain_names])
    axes[1,1].set_ylabel('Residues with missing atoms')
    axes[1,1].set_title('Missing atoms per chain')

    # Completeness statistics
    if missing_atoms_report:
        completeness_values = [record['completeness'] for record in missing_atoms_report]
        axes[1,2].hist(completeness_values, bins=20, alpha=0.7, edgecolor='black')
        axes[1,2].set_xlabel('Completeness (%)')
        axes[1,2].set_ylabel('Count')
        axes[1,2].set_title('Residue completeness distribution')
        axes[1,2].axvline(x=100, color='red', linestyle='--', alpha=0.7, label='100% complete')
        axes[1,2].legend()
    else:
        axes[1,2].text(0.5, 0.5, 'All residues complete!', ha='center', va='center',
                      transform=axes[1,2].transAxes, fontsize=12, color='green')
        axes[1,2].set_title('Residue completeness')

    plt.tight_layout()
    plt.show()

    # Renumbering function
    def renumber_atoms(df, output_filename="renumbered.pdb"):
        df = df.copy()
        df['new_atom_number'] = range(1, len(df) + 1)

        def format_pdb_line(row):
            return (f"{row['record_type']:<6}"
                    f"{row['new_atom_number']:>5} "
                    f"{row['atom_name']:>4}"
                    f"{row['alt_loc']}"
                    f"{row['residue_name']:>3} "
                    f"{row['chain_id']}"
                    f"{row['residue_number']:>4}"
                    f"{row['insertion_code']}"
                    f"   "
                    f"{row['x']:8.3f}"
                    f"{row['y']:8.3f}"
                    f"{row['z']:8.3f}"
                    f"{row['occupancy']:6.2f}"
                    f"{row['temp_factor']:6.2f}"
                    f"          "
                    f"{row['element']:>2}")

        with open(output_filename, "w") as out:
            for _, row in df.iterrows():
                out.write(format_pdb_line(row) + "\n")
            out.write("END\n")

        print(f"\n🛠️ Renumbered PDB written to: {output_filename}")

    # User prompt for renumbering
    if not non_seq_atoms.empty:
        print("\n⚠️ Atom numbers are not sequential.")
        print("To renumber the atoms, run the following in a new cell:")
        print("```python\nrenumber_atoms(df)\n```")

    # Extract sequences and export to FASTA
    print("\n" + "="*60)
    print("SEQUENCE EXTRACTION")
    print("="*60)

    # Standard amino acid three-letter to one-letter code mapping
    aa_dict = {
        'ALA': 'A', 'ARG': 'R', 'ASN': 'N', 'ASP': 'D', 'CYS': 'C',
        'GLU': 'E', 'GLN': 'Q', 'GLY': 'G', 'HIS': 'H', 'ILE': 'I',
        'LEU': 'L', 'LYS': 'K', 'MET': 'M', 'PHE': 'F', 'PRO': 'P',
        'SER': 'S', 'THR': 'T', 'TRP': 'W', 'TYR': 'Y', 'VAL': 'V',
        # Common modified residues
        'MSE': 'M',  # Selenomethionine
        'SEC': 'U',  # Selenocysteine
        'PYL': 'O',  # Pyrrolysine
    }

    def extract_sequences(df):
        """Extract protein sequences from PDB data."""
        sequences = {}

        for chain_id, chain_group in df.groupby('chain_id'):
            chain_name = chain_id if chain_id else 'blank'

            # Get unique residues in order
            residues = chain_group.groupby('residue_number').first().sort_index()

            # Filter for standard amino acids (exclude waters, ions, etc.)
            protein_residues = residues[residues['residue_name'].isin(aa_dict.keys())]

            if len(protein_residues) > 0:
                # Convert to single letter codes
                sequence = ''.join([aa_dict.get(res, 'X') for res in protein_residues['residue_name']])

                sequences[chain_name] = {
                    'sequence': sequence,
                    'length': len(sequence),
                    'residue_range': f"{protein_residues.index.min()}-{protein_residues.index.max()}",
                    'non_standard': len([res for res in protein_residues['residue_name'] if res not in aa_dict])
                }

                print(f"\n🧬 Chain {chain_name}:")
                print(f"   Length: {len(sequence)} residues")
                print(f"   Range: {protein_residues.index.min()}-{protein_residues.index.max()}")
                if sequences[chain_name]['non_standard'] > 0:
                    print(f"   Non-standard residues: {sequences[chain_name]['non_standard']} (marked as X)")
                print(f"   Sequence: {sequence}")

        return sequences

    def write_fasta(sequences, filename_base):
        """Write sequences to FASTA format."""
        fasta_filename = f"{filename_base}_sequences.fasta"

        with open(fasta_filename, 'w') as f:
            for chain_name, data in sequences.items():
                header = f">{filename_base}_chain_{chain_name}|length:{data['length']}|range:{data['residue_range']}"
                f.write(header + "\n")

                # Write sequence in 80-character lines (standard FASTA format)
                sequence = data['sequence']
                for i in range(0, len(sequence), 80):
                    f.write(sequence[i:i+80] + "\n")

        print(f"\n💾 FASTA file written to: {fasta_filename}")
        return fasta_filename

    # Extract and display sequences
    sequences = extract_sequences(df)

    if sequences:
        # Write FASTA file
        base_name = filename.replace('.pdb', '').replace('.PDB', '')
        fasta_file = write_fasta(sequences, base_name)

        print(f"\n📊 Summary:")
        print(f"   Total protein chains: {len(sequences)}")
        print(f"   Total residues: {sum(data['length'] for data in sequences.values())}")

        # Show combined statistics
        all_lengths = [data['length'] for data in sequences.values()]
        if len(all_lengths) > 1:
            print(f"   Chain lengths: {min(all_lengths)}-{max(all_lengths)} residues")

    else:
        print("\n⚠️  No protein sequences found in this PDB file.")
        print("   This might be a nucleic acid structure, small molecule, or contain only non-standard residues.")

    # Print overall missing atoms summary
    print("\n" + "="*60)
    print("MISSING ATOMS SUMMARY")
    print("="*60)

    total_incomplete = len(missing_atoms_report)
    total_residues = sum(stats['total_residues'] for stats in chain_summaries.values())

    if total_residues > 0:
        overall_completeness = ((total_residues - total_incomplete) / total_residues) * 100
        print(f"📊 Overall structure completeness: {overall_completeness:.1f}%")
        print(f"📊 Total residues analyzed: {total_residues}")
        print(f"📊 Residues with missing atoms: {total_incomplete}")

        if total_incomplete > 0:
            # Count different types of issues
            backbone_issues = sum(1 for record in missing_atoms_report if record['missing_backbone'])
            sidechain_issues = sum(1 for record in missing_atoms_report if record['missing_sidechain'])

            print(f"❌ Residues missing backbone atoms: {backbone_issues}")
            print(f"⚠️  Residues missing sidechain atoms: {sidechain_issues}")

            # Most problematic residues
            problematic_residues = defaultdict(int)
            for record in missing_atoms_report:
                problematic_residues[record['residue_name']] += 1

            if problematic_residues:
                print(f"\n🔍 Most affected residue types:")
                for res_type, count in sorted(problematic_residues.items(), key=lambda x: x[1], reverse=True)[:5]:
                    print(f"   {res_type}: {count} occurrences")
        else:
            print("✅ All amino acid residues are complete!")

    # Final recommendations
    print("\n" + "="*60)
    print("RECOMMENDATIONS")
    print("="*60)

    if total_incomplete > 0:
        print("🔧 Structure quality assessment:")

        backbone_issues = sum(1 for record in missing_atoms_report if record['missing_backbone'])
        if backbone_issues > 0:
            print(f"   ❗ CRITICAL: {backbone_issues} residues missing backbone atoms")
            print("     → This may indicate serious structural problems")
            print("     → Consider re-refinement or model validation")

        sidechain_issues = sum(1 for record in missing_atoms_report if record['missing_sidechain'])
        if sidechain_issues > 0:
            print(f"   ⚠️  WARNING: {sidechain_issues} residues missing sidechain atoms")
            print("     → Common in low-resolution structures")
            print("     → May affect binding site analysis")

        # Check for systematic truncations
        truncated_count = sum(stats['truncated_residues'] for stats in chain_summaries.values())
        if truncated_count > total_incomplete * 0.5:
            print("   🔍 High number of truncated residues detected")
            print("     → May indicate intentional mutations or modeling artifacts")
            print("     → Verify against experimental conditions")

    else:
        print("✅ Structure appears to be complete!")
        print("   → All expected atoms are present")
        print("   → Good quality for structural analysis")

print("\n✅ Analysis complete!")
````
